# Profitable Apps for the App Store and Google Play

The aim of this project is to determine the revenue of the app stroed in App Store and IOS and let the developer understand  what type of apps are likely to attract more usres

```python
from csv import reader #import the reader function from csv module

opened_file_ios = open('AppleStore.csv') #open the csv and assign the value to "opened_file_ios" variable
read_file_ios = reader(opened_file_ios) #read the "opened_file_ios" by using reader() function from csv module and assign the value to "read_file" variable
apps_data_ios = list(read_file_ios) #convert the "read_file_ios" to list and assign value to "apps_data_ios" variable
opened_file_ios.close() #close the "opened_file_ios"

opened_file_android = open('googleplaystore.csv') #open the csv and assign the value to "opened_file_andriod" variable
read_file_android = reader(opened_file_android) #read the "opened_file_andriod"
apps_data_android = list(read_file_android) #convert the "read_file_android" to list and assign value to "apps_data_android" variable
opened_file_android.close() #close the "opened_file_android"
```

```python
#function to slice data
def explore_data(dataset, start, end, rows_and_columns=False): 
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns: 
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
explore_data(apps_data_ios,1,4)
explore_data(apps_data_android,1,4)
```

```python
#Define header and data
android_header = apps_data_android[0]
android_data = apps_data_android[1:]

ios_header = apps_data_ios[0]
ios_data = apps_data_ios[1:]

#Check the incorrect data
for row in android_data:
    if len(row) != len(android_header):
        print(android_header)
        print(row)
        print("Index postion is:", android_data.index(row))
        print(f"Old len: {len(android_data)}")
        del android_data[android_data.index(row)]
        print(f"New len: {len(android_data)}")
```

```python
#Check the duplicate data
def check_dup(dataset):
    duplicate_data  = []
    unique_data = []
    for row in dataset:
        app_name = row[0]
        if app_name in unique_data:
            duplicate_data.append(app_name)
        else:
            unique_data.append(app_name)
    return (len(duplicate_data), duplicate_data[:5])
is_dup_android = check_dup(android_data)
print("Number of duplicate data for android is: {0} and the example is {1}".format(is_dup_android[0], is_dup_android[1]))

is_dup_ios = check_dup(ios_data)
print("Number of duplicate data for IOS is: {0} and the example is {1}".format(is_dup_ios[0], is_dup_ios[1]))
```

```python
#Find the highest score of the app
reviews_max = {}
for row in android_data:
    name = row[0]
    n_reviews = float(row[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews

#remove duplicate rows
android_clean = []
already_added = []
for row in android_data:
    name = row[0]
    n_reviews = float(row[3])
    
    # Check the high score of each app name and added app name to the new list
    if (n_reviews == reviews_max[name]) and (name not in already_added): 
        android_clean.append(row)
        already_added.append(name)
```

```python
def is_english(a_string):
    non_ascii = 0
    for a_char in a_string:
        if ord(a_char) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True
        
print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

```python
android_english = []
ios_english = []

for row in android_clean:
    if is_english(row[0]) == True:
        android_english.append(row)
explore_data(android_english,0,3,True)

for row in ios_data:
    if is_english(row[1]) == True:
        ios_english.append(row)
explore_data(ios_english,0,3,True)
```

```python
android_free_apps = []
ios_free_apps = []

for row in android_english:
    price = row[7]
    if price == '0':
        android_free_apps.append(row)
    
        
for row in ios_english:
    price = row[4]
    if price == '0.0':
        ios_free_apps.append(row)
        
print(len(android_free_apps))
print(len(ios_free_apps))
```

We need to find the app profiles that are successfull on App Store and Google Play be cause we need to add the app to both market.

```python
explore_data(android_free_apps, 0, 5, rows_and_columns=True)
explore_data(ios_free_apps, 0, 5, rows_and_columns=True)
```

```python
def freq_table(dataset, index):
    total_apps = 0
    table_apps = {}
    for row in dataset:
        value = row[index]
        total_apps += 1
        if value in table_apps:
            table_apps[value] += 1
        else:
            table_apps[value] = 1
            
    table_percentage = {}
    for row in table_apps:
        percentage = (table_apps[row] / total_apps) * 100
        table_percentage[row] = percentage
    return table_percentage


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
print("Prime Genre")
display_table(ios_free_apps, -5)
print('\n')
print("Categories")
display_table(android_free_apps, 1)
print('\n')
print("Genres")
display_table(android_free_apps, -4)
```

```python
ios_genre = freq_table(ios_free_apps, -5)

for genre in ios_genre:
    total = 0
    len_genre = 0
    for row in ios_free_apps:
        genre_app = row[-5]
        if genre_app == genre:
            n_ratings = float(row[5])
            total += n_ratings
            len_genre += 1
    avg_num_rating = total / len_genre
    print(genre, ':', avg_num_rating)
    
```

```python
android_genre = freq_table(android_free_apps, 1)

for category in android_genre:
    total = 0
    len_category = 0
    for row in android_free_apps:
        category_app = row[1]
        if category_app == category:
            n_install = row[5]
            n_install = n_install.replace(',', '')
            n_install = n_install.replace('+', '')
            total += float(n_install)
            len_category += 1
    avg_num_install = total / len_category
    print(category, ':', avg_num_install)
```
